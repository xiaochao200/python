#### 多分支

```python
age = input('输入年龄：')
if age < 12:
    print('你还没到12岁')
elif age < 20:
    print('你还没到20岁')
elif age < 50:
    print('你还没到50岁')
else:
    print('你超过50岁了')
```

##### 分支综合例子

```python
spam = input('请输入一个数字：')
if spam < 10:
    print('这个数小于10')
    if spam > 5:
        print('这个数还大于5')
    elif spam < 0:
        print('这个数还小于0')
    print('这个数是：',spam)
print('程序结束')
```

#### while循环

```python
spam = 0
if spam < 5:
    print('Hello, world.')
    spam = spam + 1
```

```python
spam = 0
while spam < 5:
    print('Hello, world.')
    spam = spam + 1
```

### break语句

```python
while True:
    print('Please type your name.')
    name = input()
    if name == 'xiaoming':
        break
print('Thank you!')
```

### continue语句

```python
while True:
    print('Who are you?')
    name = input()
    if name != 'J':
        continue
    print('Hello, J. What is the password?')
    password = input()
    if password == 'F':
        break
print('Access granted.')
```
